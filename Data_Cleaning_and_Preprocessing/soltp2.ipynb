### Filling missing values

```python
for column in df.columns:
    if df[column].dtype == 'object':  
        df[column].fillna(df[column].mode()[0], inplace=True)
    else:  
        df[column].fillna(df[column].mean(), inplace=True)
```

## PART 3 : Outliers

```python
df[cont_values].plot(kind='box', subplots=True, figsize=(15, 10),)
```

```python
feautures = ['alk_phosphate','bilirubin']
Q1= df[feautures].quantile(0.25) 
Q3= df[feautures].quantile(0.75)
IQR = Q3 - Q1  
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
IQR
```

```python
lower_bound
```

```python
upper_bound
```

```python
outliers_alk = df[(df['alk_phosphate'] < lower_bound['alk_phosphate']) | (df['alk_phosphate'] > upper_bound['alk_phosphate'])]
```

```python
outliers_alk
```

```python
outliers_bilirubin= df[(df['bilirubin'] < lower_bound['bilirubin']) | (df['bilirubin'] > upper_bound['bilirubin'])]
```

```python
outliers_bilirubin
```

### Handling Outliers 

```python
import numpy as np 
```

```python
df['alk_phosphate'] = np.where(df['alk_phosphate'] > upper_bound['alk_phosphate'], upper_bound['alk_phosphate'], df['alk_phosphate'])
df['alk_phosphate'] = np.where(df['alk_phosphate'] < lower_bound['alk_phosphate'], lower_bound['alk_phosphate'], df['alk_phosphate'])
```

```python
outliers_alk
```

```python
df['bilirubin'] = np.where(df['bilirubin'] > upper_bound['bilirubin'], upper_bound['bilirubin'], df['bilirubin'])
df['bilirubin'] = np.where(df['bilirubin'] < lower_bound['bilirubin'], lower_bound['bilirubin'], df['bilirubin'])
```

```python
outliers_bilirubin
```

## PART 5 : Data splitting

#### Hold-out Strategy: This strategy splits the dataset randomly into training and test sets, without considering class balance.
#### Stratified Hold-out Strategy: This strategy ensures that the class proportions in the training and test sets are similar to those in the original dataset.

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(columns=['class']) 
y = df['class']  
```

#### Hold-out strategy 


```python
X_train_ho, X_test_ho, y_train_ho, y_test_ho = train_test_split(X, y, test_size=0.3, random_state=42)
```

#### Stratified hold-out strategy


```python
X_train_sho, X_test_sho, y_train_sho, y_test_sho = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)
```

### Training samples count
